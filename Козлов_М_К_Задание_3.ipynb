```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('winequality-red.csv', sep=';')
```

```python
df.info()
```

```python
df.head()
```

```python
df["good"] = 0
condition = df["quality"] > 5
df.loc[condition, "good"] = 1
df.head(10)
```

```python
data = df['good'].value_counts()
data
```

```python
plt.figure(figsize = (10,7))

plt.pie(data, autopct='%1.1f%%')

plt.title('Доля хороших вин')
plt.legend(['хорошее', 'плохое'])

plt.show()
```

Среди представленных красных вин 53,5 процента из них оказались хорошими. Остальные - плохие.

```python
# parameters = ['alcohol', 'citric acid', 'sulphates', 'volatile acidity']
parameters = df.columns
parameters
```

```python
data_good = df[df["good"] == 1]
data_bad = df[df["good"] == 0]
```

```python
for param in parameters:
  plt.figure(figsize = (10,7))

  plt.hist(data_good[param], alpha = 0.5, label = 'good', bins=20)
  plt.hist(data_bad[param], alpha = 0.5, label = 'bad', bins=20)


  plt.title('Распределение параметра '+param)
  plt.xlabel('Значение параметра')
  plt.ylabel('Количество вин')
  plt.legend()

  plt.show()
```

Хорошим винам свойственна большая алкогольность. Ведь нет хороших вин, градус которых меньше 9. Зато есть хорошие вина, градус которых достигает 14. Фиксированная кислотность практически не влияет на качество вина. А вот чем больше летучая кислотность, общий диоксид серы, лимонной кислоты, хлоридов и чуть-чуть плотность, тем хуже красное вино. Остальные параметры не особо влияют на качество красного вина.



```python
data = df.groupby('quality')['alcohol'].agg(alco_mean = 'mean').reset_index().sort_values(by = 'quality')
data
```

```python
plt.figure(figsize = (10,7))

plt.plot(data['quality'], data['alco_mean'])
plt.title('График среднего уровня параметра "алкоголь" для вин разного качества')
plt.xlabel('Качество')
plt.ylabel('Средний уровень алкоголя')
plt.legend()

plt.show()
```

С ростом качества вина растёт и средний уровень алкоголя в красном вине. Среди слабоалкогольных вин есть много плохих.

```python
for param in parameters:
  plt.figure(figsize = (10,7))

  data = df.groupby('quality')[param].agg(mean = 'mean').reset_index().sort_values(by = 'quality')
  plt.plot(data['quality'], data['mean'])
  plt.title('График среднего уровня параметра '+param+' для вин разного качества')
  plt.xlabel('Качество')
  plt.ylabel('Средний уровень параметра '+ param)
  plt.show()
```

С увеличением среднего уровня сульфатов, фиксированной кислотности и лимонной кислоты, а также падением среднего уровня хлоридов, летучей кислотности, pH, плотности, качество вина улучшается. Больший средний уровень свободного и общего диоксида серы - у вин с оценкой 5. При этом высокий уровень остаточного сахара может быть как у плохих вин с оценкой 4, так и у хороших - с оценкой 7.
